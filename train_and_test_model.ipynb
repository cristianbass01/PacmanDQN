```python
# import module we'll need to import our custom module
from shutil import copytree

on_kaggle_server = False

if on_kaggle_server:
    kaggle_input = "/kaggle/input/dqnmodel/DQN"
    kaggle_output = "/kaggle/working/DQN"
    copytree(kaggle_input,kaggle_output)
```

```python
from DQN.DQN_model import DQNmodel
```

```python
model_type = "clipped_double"
model_detail = "hard"

model_path = f"./results/{model_type}/{model_detail}/model"
checkpoint_dir = f"./results/{model_type}/{model_detail}/checkpoints"

model = DQNmodel(
                    load_model_from=model_path,
                    checkpoint_dir=checkpoint_dir,
                    env_path="ALE/MsPacman-v5",
                    dqn_type= model_type
                )
```

```python
model.train(total_timesteps=1000000)
```

```python
logs = model.evaluate(total_episodes=1000)
```

```python
logs.save_data()
```
